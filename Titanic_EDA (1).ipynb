```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset into a dataframe
titanic_data = pd.read_csv("week9_titanic_eda_data.csv")
# Display the first few rows of the dataset
print("week9_titanic_eda_data.csv.head()")
print(titanic_data.head())
```

```python
#Using info to summarize the dataset
print("info of the dataset including data types and non-null counts:", titanic_data.info)
```

```python
#Using Describe to summarize the dataset
print("Description of the Dataset", titanic_data.describe())
```

3. What is the distribution of survival (Survived column)?

```python
print("Distribution of survival (0=Did not survive, 1=Survived")
print(titanic_data['Survived'].value_counts())
```

The number of passengers that survived is 2496


4. Use .value_counts() and a bar chart to show the count of survivors vs non-survivors.

```python
print("A bar chart delineating survivors and non-survivors")
survival_counts=titanic_data['Survived'].value_counts()
#Creating a bar chart
survival_counts.plot(kind='bar',color=['red','green'])
#Add chart labels and title
plt.title(" Survivors vs Non-survivors")
plt.xlabel("Survived(0=No, 1=Yes)")
plt.ylabel("Number of passengers")
plt.xticks([0,1],["Did Not Survive","Survived"],rotation=0)
#Show plot
plt.tight_layout()
plt.show()
```

5. What is the distribution of passenger classes (Pclass)?

```python
print("The distribution of passenger classes (Pclass)")
print(titanic_data['Pclass'].value_counts())
```

6. Use .value_counts() and a pie or bar chart to visualize it.

```python
print("Use value_counts() and a pie or bar chart to visualize it.")
# Pie chart
class_counts = titanic_data['Pclass'].value_counts()
class_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'], labels=['First Class', 'Second Class', 'Third Class'])
plt.ylabel('')
plt.title('Passenger Class Distribution')
plt.show()
```

7. What is the age distribution of the passengers?
8. Plot a histogram or KDE plot of the Age column.
9. Include handling of missing values.


```python
#Checking missing values
print(f"Missing Age values: {titanic_data['Age'].isnull().sum()}")
# Age distribution
print("\nAge distribution of passengers:")
print(titanic_data['Age'].value_counts())
# Histogram
plt.hist(titanic_data['Age'], bins=30, edgecolor='black', color='#66b3ff')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

10. What is the relationship between Age and Survival?
11. Use boxplots or violin plots grouped by survival status.


```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

12. What is the gender breakdown of the passengers?
13. Use .value_counts() and visualize with a bar chart.


```python
Gender_breakdown = titanic_data['Sex'].value_counts()
print(Gender_breakdown)
Gender_breakdown.plot(kind='bar', color=['red', 'green'])
plt.title("Gender Breakdown of the Passengers")
plt.xlabel("Gender")
plt.ylabel("Number of passengers")
plt.xticks([0, 1], ["Female", "Male"], rotation=0)
plt.tight_layout()
plt.show()
```

14. a. Then cross-tabulate with Survived to analyze survival by gender.

```python
gender_survival = pd.crosstab(titanic_data['Sex'], titanic_data['Survived'])
print(gender_survival)
# Stacked bar chart of survival by gender
gender_survival.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(6,4))
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

15b What are the most common embarkation ports?
15c Use .value_counts() on Embarked and visualize it.

```python
# What are the most common embarkation ports?
print("Most common embarkation ports:")
embark_counts = titanic_data['Embarked'].value_counts()
print(embark_counts)

# Visualize the embarkation ports
embark_counts.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Embarkation Port Distribution')
plt.xlabel('Port of Embarkation')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1, 2], ['Queenstown', 'Southampton', 'Cherbourg'], rotation=0)
plt.show()
```
